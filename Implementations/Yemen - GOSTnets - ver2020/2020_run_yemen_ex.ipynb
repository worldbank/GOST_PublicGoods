````python
import numpy as np
````

````python
OD = OD.replace([np.inf, -np.inf], np.nan)
````

## A very simple Accessibility Analysis

````python
closest_facility_per_origin = OD.min(axis = 1)
````

````python
closest_facility_per_origin.index
````

````python
results = pd.DataFrame([closest_facility_per_origin.index,closest_facility_per_origin]).transpose()
````

````python
results
````

````python
results.columns = ["NN", "travel_time_to_closest_facility"]
````

````python
results
````

## Now time to merge origins

````python
snapped_origins_2018_copy = snapped_origins_2018
````

````python
output = pd.merge(snapped_origins_2018_copy, results, on = "NN")
````
