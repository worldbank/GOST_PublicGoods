Using GOSTnets you will build a really small network from an osm pbf file. You will add time to the network edges. You will then convert the graph to an edge GDF, and finally export as a shapefile. 

```python
import geopandas as gpd
import pandas as pd
import os, sys
```

```python
sys.path.append(r"C:\repos\GOSTnets")
import GOSTnets as gn
```

```python
from GOSTnets.load_osm import *
```

```python
Reykjavik = OSM_to_network(r"input/rey_small.osm.pbf")
```

```python
Reykjavik.generateRoadsGDF(verbose = False)
```

```python
Reykjavik.roadsGPD
```

```python
Reykjavik.initialReadIn()
```

```python
gn.example_edge(Reykjavik.network)
```

```python
G_Reykjavik_time = gn.convert_network_to_time(Reykjavik.network, 'length', graph_type = 'drive', road_col = 'infra_type', output_time_col = 'time', speed_dict = None, walk_speed = 4.5, factor = 1000, default = None)
```

```python
gn.example_edge(G_Reykjavik_time)
```

```python
#gn.save(G_Reykjavik_time, 'G_Reykjavik_time', '', pickle=False, edges=True, nodes=False)
```

```python
G_Reykjavik_time_gdf = gn.edge_gdf_from_graph(G_Reykjavik_time)
```

```python
# can't have two geometry columns, so drop the 'Wkt' one
G_Reykjavik_time_gdf = G_Reykjavik_time_gdf.drop(['Wkt'],axis=1)
```

```python
G_Reykjavik_time_gdf
```

```python
G_Reykjavik_time_gdf.to_file("G_Reykjavik_time_gdf.shp")
```
